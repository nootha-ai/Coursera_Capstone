##### Extract only Zip,Latitude and Longitude columns

```python
df_zip_lat_long=df_zipcodes.drop(['City','State','Timezone','Daylight savings time flag','geopoint'],axis=1)

df_zip_lat_long
```

```python
df_zip_neighborhood =pd.DataFrame(df_neighbor.groupby(['ZipCode']).Neighborhood.agg([('Neighborhood', ', '.join)]))
df_zip_neighborhood.reset_index(inplace=True)
df_zip_neighborhood
```

```python
print('The shape of the dataset is:',df_zip_neighborhood.shape)
```

```python
df_zip_lat_long.rename(columns={'Zip': 'ZipCode'}, inplace=True)
df_zip_lat_long.head()
```

```python
df_neighborhoods_coordinates = pd.merge(df_zip_neighborhood, df_zip_lat_long, on='ZipCode')
df_neighborhoods_coordinates.head()
```

```python
#Create list with the zipcodes (to be used later)
zipcodes = df_neighborhoods_coordinates['ZipCode'].unique().tolist()
```

```python
lat_sacramento = df_neighborhoods_coordinates['Latitude'].mean()
lon_sacramento = df_neighborhoods_coordinates['Longitude'].mean()
print('The geographical coordinates of Toronto are {}, {}'.format(lat_sacramento, lon_sacramento))
```

```python
zipcode_color = {}
for zipcode in zipcodes:
    zipcode_color[zipcode]= '#%02X%02X%02X' % tuple(np.random.choice(range(256), size=3)) 
```

```python
map_sacramento = folium.Map(location=[lat_sacramento, lon_sacramento], zoom_start=12)

# add markers to map
for lat, lng, zipcode, neighborhood in zip(df_neighborhoods_coordinates['Latitude'], 
                                           df_neighborhoods_coordinates['Longitude'],
                                           df_neighborhoods_coordinates['ZipCode'],
                                           df_neighborhoods_coordinates['Neighborhood']):
    label_text = str(zipcode) + ' - ' + neighborhood
    label = folium.Popup(label_text)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=zipcode_color[zipcode],
        fill_color=zipcode_color[zipcode],
        fill_opacity=0.7).add_to(map_sacramento)  
    
map_sacramento
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
   
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
sacramento_venues = getNearbyVenues(names=df_neighborhoods_coordinates['Neighborhood'],
                                latitudes=df_neighborhoods_coordinates['Latitude'],
                                longitudes=df_neighborhoods_coordinates['Longitude'])
```

```python
to_onehot = pd.get_dummies(sacramento_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
to_onehot['Neighborhoods'] = sacramento_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [to_onehot.columns[-1]] + list(to_onehot.columns[:-1])
to_onehot = to_onehot[fixed_columns]

print(to_onehot.shape)
to_onehot.head()
```

```python
to_grouped = to_onehot.groupby(["Neighborhoods"]).mean().reset_index()

print(to_grouped.shape)
to_grouped
```

```python
len(to_grouped[to_grouped["Grocery Store"] > 0])
```

```python
to_GroceryStore = to_grouped[["Neighborhoods","Grocery Store"]]
```

```python
to_GroceryStore
```

```python
from sklearn.cluster import KMeans
toclusters =3

to_clustering = to_GroceryStore.drop(["Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=toclusters, random_state=1)
kmeans.fit_transform(to_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
to_merged = to_GroceryStore.copy()

# add clustering labels
to_merged["Cluster Labels"] = kmeans.labels_
```

```python
to_merged.rename(columns={"Neighborhoods": "Neighborhood"}, inplace=True)
to_merged
```

```python
to_merged = to_merged.join(sacramento_venues.set_index("Neighborhood"), on="Neighborhood")

print(to_merged.shape)
to_merged
```

```python
to_merged.sort_values(["Cluster Labels"], inplace=True)
to_merged
```

```python
map_clusters = folium.Map(location=[lat_sacramento, lon_sacramento],zoom_start=14)

# add markers to the map
markers_colors={}
markers_colors[0] = 'green'
markers_colors[1] = 'red'
markers_colors[2] = 'blue'
markers_colors[3] = 'yellow'
markers_colors[4] = 'cyan'
markers_colors[5] = 'black'
for lat, lon, cluster in zip(to_merged['Neighborhood Latitude'], to_merged['Neighborhood Longitude'], to_merged['Cluster Labels']):
    
    
    folium.CircleMarker(
        [lat, lon],
        radius=5,
       
        color =markers_colors[cluster],
        fill_color=markers_colors[cluster],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
